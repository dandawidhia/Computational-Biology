```python
from Bio.Seq import Seq

SeqA = ("GACGATCGATGCTAGCCAC")
```

```python
def GCComposition(seq):
    countG = seq.count("G")
    countC = seq.count("C")
    return (countG + countC) / len(seq) * 100

print("GC Composition", round(GCComposition(SeqA), 2), "%")
print("AT Composition", 100 - round(GCComposition(SeqA), 2), "%")
```

```python
from Bio.SeqUtils import gc_fraction

print(round(gc_fraction(SeqA)), "%")
```

```python
# TM = 2(A+T) + 4(G+C)

def WallaceRule(seq):
    countA = seq.count("A")
    countT = seq.count("T")
    countG = seq.count("G")
    countC = seq.count("C")

    # buat dupe ke bawah, shortcut -> alt + shift + down key arrrow
    return 2 * (countA + countT) + 4 * (countG + countC)

seqC = Seq('GAGCTAGCGQATGCAGCAGCAG')
print(WallaceRule(seqC))
```

```python
# Global Alignment
# makin tinggi angka semakin mirip
import Bio
from Bio.Seq import Seq
from Bio import pairwise2
from Bio.pairwise2 import format_alignment
```

```python
# For Scoring
# xx = default: tiap match score +1
# mx -> Custom: match and mismatch
# ms -> Custom: match, mismatch, gap start, dan gap extend

SeqA = Seq("GGTTTTG")
SeqB = Seq("GG")
globalAlignments = pairwise2.align.globalxx(SeqA, SeqB)
print (globalAlignments)
# for all Alignment in globalAlignments: print(globalAlignments)

for alignment in globalAlignments:
  print(format_alignment(*alignment)) # for unpack, start, end, score
```

```python
# Local Alignment
LocalAlignments = pairwise2.align.localxx(SeqA, SeqB)
for alignment in LocalAlignments:
  print(format_alignment(*alignment)) # for unpack, start, end, score
```

```python
# Custom Scoring

SeqC = Seq("GCATG")
SeqD = Seq("GCCTG")

# Scoring
# Match = +1 score
# Missmatch = -1 score
# Gap = -0.5 score
# Gap Extend = -0.1

customGlobalAlignments = pairwise2.align.globalms(SeqC, SeqD, 1, -1, -0.5, -0.1)
print(customGlobalAlignments)

for alignment in customGlobalAlignments :
  print(format_alignment(*alignment))
```

```python
# Hamming Distance
# dihitung sesuai panjang jika kurang maka 0

def hammingDistance(seq1, seq2):
  score = 0
  for x, y in zip(seq1, seq2):
    if x != y:
      score += 1
  return score


SeqE = Seq("GGGGG")
SeqF = Seq("AAAAA")

print(hammingDistance(SeqE, SeqF))
```

```python
Seqx = x.seq.replace("N", "")
print('new SeqX =', Seqx)
```

# 1. SEQUENCE MANIPULATION

```python
# Menampilkan panjang Urutan (length) X dan Y.

print("Lenght seq x: ", len(Seqx))
print("Lenght seq y: ", len(Seqy))

# Menampilkan total panjang Urutan (length) X dan Y.
print("Total seq: ", len(Seqx + Seqy))
```

```python
# Temukan indeks kemunculan pertama kodon 'CGC'
# FIND

print("Index codon x: ", Seqx.find('CGC'))
print("Index codon y: ", Seqy.find('CGC'))
```

```python
# Melakukan slicing
# array[start:stop:step]

# Take the first 10 nucleotides from Sequence X and the last 5 nucleotides from Sequence X.
Seqx1 = Seqx[:10] + Seqx[-5:]

# Take the first 20 nucleotides from Sequence Y and the last 3 nucleotides from Sequence Y.
Seqy1 = Seqy[:20] + Seqy[-3:]

# Combine both of the string above and reverse it to a new sequence, Sequence Z.
Seqz = Seqx1 + Seqy1
Seqz = Seqz[::-1]
print("Seqz: ", Seqz)
```
